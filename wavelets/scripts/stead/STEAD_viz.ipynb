```python
import pickle5
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
with open('stead_samples.pickle', 'rb') as handle:
    samples = pickle5.load(handle)
vals = list(samples.values())
```

```python
random.choice(vals)
```

```python
def plot_randos(sample):
    
    fig = plt.figure(figsize=(15, 10))
    axs = fig.subplots(2, 1, sharex=True, gridspec_kw={"hspace": 0, "height_ratios": [3, 1]})
    axs[0].plot(sample["data"][0], label='Z')
    axs[0].plot(sample["data"][1], label='N')
    axs[0].plot(sample["data"][2], label='E')
    axs[0].legend()
    axs[1].plot(sample["truth"][0], label='P-Pick')
    axs[1].plot(sample["truth"][1], label='S-Pick')
    axs[1].plot(sample["truth"][2], label='Noise')
    axs[1].legend()
    plt.show()
```

```python

for i in range(25):
    sample = random.choice(vals)
    plot_randos(sample)
```
